**Because in the task, we need to identify people with mental health issues, I have decided to leave only the "treatment" column to show the problems. This is the only column that definitely shows if a person has a problem or not.**

```python
df_copy = df.copy()
```

```python
df = df[['Age', 'Gender', 'Country', 'treatment', 'tech_company']]
```

```python
df = df[df['tech_company'] == 'Yes']
```

```python
df['Gender'].unique()
```

```python
def gender_map(gender):
  gender = str(gender).strip().lower()
  if gender in ['male', 'm', 'man', 'male-ish', 'cis male', 'male (cis)', 'msle',
                 'mal', 'maile', 'make', 'male ', 'cis man', 'malr', 'mail']:
    return 'Male'
  elif gender in ['female', 'f', 'woman', 'femail', 'cis female', 'cis-female/femme',
                  'female ', 'femake', 'female (cis)']:
    return 'Female'
  else:
    return 'Other'
```

```python
df['Gender_categories'] = df['Gender'].apply(gender_map)
```

```python
df['Gender_categories'].value_counts()
```

```python
df.head()
```

```python
df.drop(columns=['Gender'], inplace=True)
```

```python
df['Age'].unique()
```

```python
df['Age'].value_counts()
```

```python
df = df[df['Age'].apply(lambda x: str(x).isdigit() and 16 < int(x) < 100)]
```

```python
df['treatment'].value_counts()
```

```python
df['treatment'] = df['treatment'].map({'Yes': 1, 'No': 0})
```

```python
country_counts = df['Country'].value_counts()
```

```python
df['Country'].unique()
```

```python
rare_countries = country_counts[country_counts < 5].index
```

```python
df['Country'] = df['Country'].replace(rare_countries, 'Other')
```

**Need to create age groups**

```python
bins = [0, 18, 29, 40, 50, 100]
labels = ['<18', '18-29', '30-40', '41-50', '>50']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

**Find the mean in each age group**

```python
age_distribution = df.groupby('Age_Group', observed=False)['treatment'].mean()
age_distribution
```

The same (finding the mean in each Gender)

```python
gender_distribution = df.groupby('Gender_categories', observed=False)['treatment'].mean()
gender_distribution.sort_values(ascending=False)
```

```python
country_counts  = df.groupby('Country', observed=False)['treatment'].mean()
country_counts.sort_values(ascending=False)
```

# CREATING THE GRAPHS

```python
age_distribution.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Treatment Rate')
plt.title('Treatment Rate by Age Group')
plt
```

**Judging by the diagram, we can say that mental issues are more common in the group of people over the age of 50**

```python
gender_distribution.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Treatment Rate')
plt.title('Treatment Rate by Gender')
plt
```

**Judging by the graph of a person with non-standard gender identification, women suffer from mental problems more often than men. But since we assessed the presence of mental problems on an objective basis (treatment), it is difficult to say for sure, since it is possible that women are more likely to seek psychological help (as well as LGDT+).**

```python
country_counts = country_counts.sort_values()
country_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Treatment Rate')
plt.title('Treatment Rate by Country')
plt.grid(axis='y', linestyle='--', alpha=0.8)
plt
```

**Highest rate: Australia has the highest treatment rate, exceeding 0.6, indicating a higher average number of people suffering from mental problems or the availability of treatment in the region.**

**Lowest rate: France has the lowest treatment rate, less than 0.2, which indicates that in France, tech industry workers suffer from mental problems less often or seek help less often**
