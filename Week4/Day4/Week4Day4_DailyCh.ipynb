```python
Q1 = df['salary'].quantile(0.25)  # 25-й перцентиль
Q3 = df['salary'].quantile(0.75)  # 75-й перцентиль
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['salary_capped'] = df['salary'].clip(lower_bound, upper_bound)
```

```python
df.describe()
```

```python
scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary_capped']])
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
oh_encoder = OneHotEncoder(sparse_output=False, drop='first')
job_title_encoded = oh_encoder.fit_transform(df[['job_title']])
job_title_encoded_df = pd.DataFrame(job_title_encoded, columns=oh_encoder.get_feature_names_out(['job_title']))

job_type_encoded = oh_encoder.fit_transform(df[['job_type']])
job_type_encoded_df = pd.DataFrame(job_type_encoded, columns=oh_encoder.get_feature_names_out(['job_type']))
```

```python
df = pd.concat([df, job_title_encoded_df], axis=1)
df = pd.concat([df, job_type_encoded_df], axis=1)
```

```python
oh_encoder = OneHotEncoder(sparse_output=False, drop='first')
salary_currency_encoded = oh_encoder.fit_transform(df[['salary_currency']])
salary_currency_encoded_df = pd.DataFrame(salary_currency_encoded, columns=oh_encoder.get_feature_names_out(['salary_currency']))
```

```python
df = pd.concat([df, salary_currency_encoded_df], axis=1)
```

```python
df = pd.concat([df, job_title_encoded_df, job_type_encoded_df], axis=1)
```

```python
df.drop(columns=['job_title', 'job_type'], inplace=True)
```

```python
le = LabelEncoder()
df['experience_level_encoded'] = le.fit_transform(df['experience_level'])
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df['location'].unique()
```

```python
top_locations = df['location'].value_counts().nlargest(20)
```

```python
df['location_grouped'] = df['location'].apply(lambda x: x if x in top_locations else 'Other')
```

```python
df['location_grouped'].value_counts()
```

```python
columns_to_drop = ['location', 'salary_currency', 'salary','experience_level','location_grouped','salary_capped']
df_reduced = df.drop(columns=columns_to_drop)
```

```python
df_reduced.head()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_reduced)
```

```python
pca = PCA(n_components=0.95)
pca_result = pca.fit_transform(scaled_data)
```

```python
pca_columns = [f'PC{i+1}' for i in range(pca_result.shape[1])]
df_pca = pd.DataFrame(pca_result, columns=pca_columns)
```

```python
df_pca.head()
```

```python
df_reduced= pd.concat([df, df_pca], axis=1)
```
