```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)
```

```python
# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
#Rotate an image by 90 degrees
rotated_image = original_image.rotate(90)
plt.imshow(rotated_image)
```

```python
#Rotate an image by 50 degrees
rotated_image = original_image.rotate(50)
plt.imshow(rotated_image)
```

```python
#Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
mirrored_image = ImageOps.mirror(original_image)
plt.imshow(mirrored_image)
```

```python
flipped_image = ImageOps.flip(mirrored_image)
plt.imshow(flipped_image)
```

```python
#Zoom in on an image (scale by 1.2x) using .resize(...)

width, height = original_image.size
new_width, new_height = int(width * 1.2), int(height * 1.2)
zoomed_image = original_image.resize((new_width, new_height))
plt.imshow(zoomed_image)
```
