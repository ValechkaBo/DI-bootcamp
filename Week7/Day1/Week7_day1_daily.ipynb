```python
import bs4
import requests
import pandas as pd
```

```python
url = 'https://github.com/topics'

r = requests.get(url)

if r.status_code == 200:
    print('Success')

    r.encoding = r.apparent_encoding
    html_content = r.text
    print(html_content[:100])

    with open('webpage.html',  "w", encoding="utf-8") as f:
        f.write(html_content)
        print('Done')

else:
    print('Failed')




```

```python
soup = bs4.BeautifulSoup(html_content, 'html.parser')

soup.prettify()

```

```python
title = soup.find('title')
print(title)
```

```python
list_title =[]
list_desc = []

def get_topic_titles(soup):
    topic = soup.find_all('p', class_='f3 lh-condensed mb-0 mt-1 Link--primary')
    topic_titles = []
    for i in topic:
        topic_titles.append(i.text)
    return topic_titles
    list_title.append(topic_titles)
def get_topic_descs(soup):
    desc = soup.find_all('p', class_='f5 color-fg-muted mb-0 mt-1')
    topic_descs = []
    for i in desc:
        topic_descs.append(i.text.strip())
    return topic_descs
    list_desc.append(topic_descs)

list_title = get_topic_titles(soup)
list_desc = get_topic_descs(soup)

df = pd.DataFrame( columns=['Title'], index=None)
df['Title'] = list_title
df['Description'] = list_desc
df
```
