```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.random.randint(-5,36, size=(10,12))
temp
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
city = ['City1','City2','City3','City4','City5','City6','City7','City8','City9','City10']
df = pd.DataFrame(temp, columns=months, index=city)
df
```

```python
mean_city = df.mean(axis=1)
mean_city
```

```python
mean_city.idxmax() #average max
```

```python
mean_city.idxmin() #average min
```

```python
city_min = df.min(axis=1) ##min temp in each city
city_min
```

```python
city_min.idxmin()
```

```python
city_max = df.max(axis=1) #max temp in each city
```

```python
city_max.idxmax()
```

**Visualisation **

```python
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(10,6))
for city in df.index:
  plt.plot(df.columns, df.loc[city], label=city)
plt.title('Temperature Trends for Cities')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.legend(title="Cities", bbox_to_anchor=(1.05, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
