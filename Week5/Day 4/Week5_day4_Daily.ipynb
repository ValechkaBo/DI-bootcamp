**Create a histogram showing the distribution of car prices.**

```python
fig = px.histogram(df, x='price', nbins=30, title='Distribution of Car Prices', labels={'price': 'Price'})

fig.update_layout(bargap=0.1)
fig.update_xaxes(title_text='Price', tickfont=dict(size=12))
fig.update_yaxes(title_text='Count', tickfont=dict(size=12))

#to separate bars and more well looking
fig.update_traces(marker=dict(
    color='#C71585',
    line=dict(color='black', width=1),
    opacity=0.7))

fig.update_layout(       #making title and axeses name more readable
    title="Distribution of Car Prices",
    xaxis_title="Price",
    yaxis_title="Count",
    font=dict(
        family="Arial, sans-serif",
        size=20,
        color="black"
    )
)
fig.show()


```

# Create a box plot showing price distribution by brand (top 10 brands).

```python
df['brand'].value_counts()
```

```python
df_brands = df['brand'].value_counts().head(10)    #find top 10
df_brands
```

```python
df_top_brands = df[df['brand'].isin(df_brands.index)]
df_top_brands
```

```python
fig = px.box(df_top_brands, x='brand', y='price', title='Price Distribution by Brand (Top 10 Brands)', labels={'brand': 'Brand', 'price': 'Price'})
fig.update_layout(
    font=dict(
        family="Arial, sans-serif",
        size=20,
        color="black"
    )
)
fig.update_layout(
    title={
        'text': "Price Distribution by Brand (Top 10 Brands)",
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)


fig.show()
```

# Create an interactive scatter plot of Price vs. Mileage, color-coded by Condition.

```python
df = df.drop(df[df['mileage'] > 400000].index)
```

```python
fig = px.scatter(df, x='mileage', y='price', color='condition', title='Price vs. Mileage (Color-coded by Condition)', labels={'mileage': 'Mileage', 'price': 'Price'})
fig.update_layout(
    font=dict(
        family="Arial, sans-serif",
        size=18,
        color="black"
    )
)
fig.update_layout(
    title={
        'text': "Price vs. Mileage (Color-coded by Condition)",
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)
```

# Optional : create a heatmap of average prices by brand and condition

```python
mean_price_by_brand_condition = df.groupby(['brand', 'condition'])['price'].mean().reset_index()
```

```python
heatmap_data = mean_price_by_brand_condition.pivot(index='brand', columns='condition', values='price')
```

```python
fig = px.imshow(heatmap_data,
                labels=dict(x="Condition", y="Brand", color="Average Price"),
                x=heatmap_data.columns,
                y=heatmap_data.index,
                color_continuous_scale='Viridis')
fig.update_layout(
    title={
        'text': "Average Price by Brand and Condition",
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    }
)
fig.update_layout(
    font=dict(
        family="Arial, sans-serif",
        size=18,
        color="black"
    )
)
fig.update_xaxes(tickangle=-45)
fig.show()
```
