While the Adele and Chinstrap penguins go neck and neck, we can still notice that there are some individuals of the Adele species that are larger in mass and some of the Chinstrap species with longer fins.

I hope Gentoo doesn't hurt its younger brothers!

Heatmap of Correlations:
Plot a heatmap to visualize the correlations between numerical variables.

```python
df.head()
```

```python
df.rename(columns={                           #renamed to not to deal with that anymore
    'bill_length_mm': 'Bill Length (mm)',
    'bill_depth_mm': 'Bill Depth (mm)',
    'flipper_length_mm': 'Flipper Length (mm)',
    'body_mass_g': 'Body Mass (g)'
}, inplace=True)
```

```python
cor_matrix = df[['Bill Length (mm)', 'Bill Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)']].corr()   #find the correlation
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm', linewidths=1, fmt=".2f")
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.title('Heatmap of Correlations')
```

1. There is a strong positive correlation between flipper length and body mass. This is logical, as bigger birds have larger flippers.
2. There is also a correlation between bill size and flipper size. Penguins with larger flippers tend to have larger bills.
3. Additionally, there is a positive correlation between bill length and body mass.
Overall, we can conclude that the heavier the bird, the larger its flippers and bill.

But there is a negative correlation between the bill depth  and other characteristics,
especially with a long flipper and body weight. It is possible that smaller penguins have a deeper bill

# Boxplots for Measurements:
Generate boxplots for flipper length, body mass, and bill measurements, segmented by species.

```python
#for the better visualization I added some manual setting

sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

sns.boxplot(ax=axes[0, 0],data=df, x='species', y='Flipper Length (mm)', hue='species', palette='pastel')
axes[0, 0].set_title('Flipper Length by Species')
axes[0, 0].set_xlabel('Species')
axes[0, 0].set_ylabel('Flipper Length (mm)')

sns.boxplot(ax=axes[0, 1], data=df, x='species', y='Bill Length (mm)',hue='species', palette='pastel')
axes[0, 1].set_title('Bill Length by Species')
axes[0, 1].set_xlabel('Species')
axes[0, 1].set_ylabel('Bill Length (mm)')

sns.boxplot(ax=axes[1, 0], data=df, x='species', y='Bill Depth (mm)',hue='species', palette='pastel')
axes[1, 0].set_title('Bill Depth by Species')
axes[1, 0].set_xlabel('Species')
axes[1, 0].set_ylabel('Bill Depth (mm)')

sns.boxplot(ax=axes[1, 1], data=df, x='species', y='Body Mass (g)',hue='species', palette='pastel')
axes[1, 1].set_title('Body Mass by Species')
axes[1, 1].set_xlabel('Species')
axes[1, 1].set_ylabel('Body Mass (g)')

plt.tight_layout() # for good looking
plt.subplots_adjust(hspace=0.4, wspace=0.4) #add space between graphs

plt.suptitle('Boxplots for Measurements')

plt.tight_layout()
```

We can say that the penguins at Palmer Station differ in appearance.

1. The most obvious thing is that the Gentoo species is larger than the rest. The carcass of these penguins is larger and the flippers are longer.  

2. Adelie penguins have shorter flippers and a smaller but deeper bill.

3. Chinstrap occupy an average position in terms of characteristics and have similar indicators to Adele , but their bill is usually much longer

4. It is worth noting that considering the sample divided by species, we find that there are unique outliers among individuals according to different characteristics. When analyzing the overall sample, these observations were not obvious

Histograms of Measurements: Create histograms to show the distribution of key measurements (bill length, bill depth, flipper length).

```python
fig, axes = plt.subplots(2, 2, figsize=(15,12))

sns.histplot(ax=axes[0, 0], data=df, x='Bill Length (mm)', hue='species', palette='pastel', kde=False, multiple='stack')
axes[0, 0].set_title('Bill Length Distribution')
axes[0, 0].set_xlabel('Bill Length (mm)')
axes[0, 0].set_ylabel('Count')

sns.histplot(ax=axes[1, 0], data=df, x='Bill Depth (mm)', hue='species', palette='pastel', kde=False, multiple='stack')
axes[1, 0].set_title('Bill Depth Distribution')
axes[1, 0].set_xlabel('Bill Depth (mm)')
axes[1, 0].set_ylabel('Count')

sns.histplot(ax=axes[0, 1], data=df, x='Flipper Length (mm)',hue='species', palette='pastel', kde=False, multiple='stack')
axes[0, 1].set_title('Flipper Length Distribution')
axes[0, 1].set_xlabel('Flipper Length (mm)')
axes[0, 1].set_ylabel('Count')

sns.histplot(ax=axes[1, 1], data=df, x='Body Mass (g)', hue='species', palette='pastel', kde=False, multiple='stack')
axes[1, 1].set_title('Body Mass Distribution')
axes[1, 1].set_xlabel('Body Mass (g)')
axes[1, 1].set_ylabel('Count')

plt.tight_layout(pad=2.2, h_pad=2.2, w_pad=2.2)

plt.suptitle('Histograms of Measurements')

plt.tight_layout
```

# Based on the histograms, here are some short conclusions:

1. Bill Length:

Adelie has shorter bills (35-45 mm).

Gentoo and Chinstrap have longer bills (45-55 mm).

2. Flipper Length:

Gentoo has the longest flippers (200-230 mm).
Adelie has the shortest flippers (170-200 mm).
3. Bill Depth:

Adelie has deeper bills (16-20 mm).
Gentoo has shallower bills (14-17 mm).
4. Body Mass:

Gentoo is the heaviest (4500-6000 g).
Adelie is lighter (3000-4000 g).
Chinstrap is between Adelie and Gentoo.


**Summary:**

Gentoo is the biggest species with longer flippers and heavier body mass.
Adelie has shorter bills and smaller body size.
Chinstrap is in the middle, with characteristics between Adelie and Gentoo
