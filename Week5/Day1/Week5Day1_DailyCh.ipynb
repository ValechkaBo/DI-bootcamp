```python
df_country = pd.read_csv('countries_of_the_world.csv')
df_country.head()
```

```python
df.columns = df.columns.str.strip()
df_country.columns = df_country.columns.str.strip()
```

```python
df = df.rename(columns={'Country or region': 'Country'})
```

```python
df['Country'] = df['Country'].str.strip()
df_country['Country'] = df_country['Country'].str.strip()
```

```python
df = df.merge(df_country[['Country', 'Region']], on='Country', how='left')
```

```python
df.head()
```

```python
df['Region'].unique()
```

```python
df['Region'].isnull().sum()
```

```python
df = df.drop(['Region_y', 'Region_x'], axis=1)
```

```python
missing_rows = df[df['Region'].isnull()]
print(missing_rows)
```

```python
df['Region'].unique()
```

```python
region_mapping = {
    'Kosovo': 'EASTERN EUROPE',
    'South Korea': 'ASIA (EX. NEAR EAST)',
    'Northern Cyprus': 'NEAR EAST',
    'Montenegro': 'EASTERN EUROPE',
    'Bosnia and Herzegovina': 'EASTERN EUROPE',
    'North Macedonia': 'EASTERN EUROPE',
    'Ivory Coast': 'SUB-SAHARAN AFRICA',
    'Congo (Brazzaville)': 'SUB-SAHARAN AFRICA',
    'Palestinian Territories': 'NEAR EAST',
    'Gambia': 'SUB-SAHARAN AFRICA',
    'Congo (Kinshasa)': 'SUB-SAHARAN AFRICA',
    'Myanmar': 'ASIA (EX. NEAR EAST)',
    'Central African Republic': 'SUB-SAHARAN AFRICA',
    'South Sudan': 'SUB-SAHARAN AFRICA'
}
```

```python
df['Region'] = df['Region'].fillna(df['Country'].map(region_mapping))
```

```python
regions = df['Region'].unique()
colors = plt.cm.tab10(range(len(regions)))
color_map = dict(zip(regions, colors))
plt.figure(figsize=(10, 6))
for region in regions:
    plt.scatter(df[df['Region'] == region]['GDP per capita'], df[df['Region'] == region]['Healthy life expectancy'], label=region, color=color_map[region])


plt.xlabel('GDP per Capita')
plt.ylabel('Healthy Life Expectancy')
plt.title('Correlation Between GDP per Capita and Healthy Life Expectancy by Region')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
